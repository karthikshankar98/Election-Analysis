# Loksabha Elections 2014 

```r
install.packages('ggplot2')
library(ggplot2)

install.packages('dplyr')
library(dplyr)
```

```r
df <- read.csv('LS-2014.csv')
```

```r
str(df)
```

```r
summary(df)
```

```r
ggplot(df,aes(Cand.No,Total.recorded)) + geom_point(color='red', size=2) + scale_x_continuous(breaks = seq(min(1), max(29), by = 1))
```

```r
ggplot(df,aes(Booth.No,Meenakshi.Lekhi)) + geom_point(color = "blue") 
```

It can be seen from the above graph that Canditate No. 3 (Meenakshi Lekhi) got the highest votes

# Elections 2020 Prediction 

For prediction of the elections in 2020, the data from Vidhansabha Elections of 2013 is taken as an example

```r
install.packages('e1071')
library(e1071)

install.packages('caTools')
library(caTools)
```

```r
df <- read.csv('VS-2013.csv')
```

```r
str(df)
```

```r
df <- select(df,-'Total.votes',-'Cand.No',)
```

```r
sample <- sample.split(df$Saurabh.Bharadwaj, SplitRatio = 0.7) 
train <- subset(df, sample=TRUE) 
test <- subset(df, sample=FALSE)
```

```r
model <- svm(Saurabh.Bharadwaj ~ ., data = train)
```

```r
summary(model)
```

```r
pred <- fitted(model)
pred
```

```r
table(pred, test$Saurabh.Bharadwaj)
```
